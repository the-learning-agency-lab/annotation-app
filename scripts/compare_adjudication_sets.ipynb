```python
import pandas as pd

subset1a = pd.read_json("../inputs/ume-adjudication/subset-1a.jsonl", lines=True)
examples = pd.read_json("../data/ume-final.jsonl", lines=True)
examples["_timestamp"] = pd.to_datetime(examples["_timestamp"], unit='s')
subset1b = pd.read_csv("../data/Mar6data_ToAdjudicate.csv")
subset1a
```

```python
a = set(subset1a["idx"])
b = set(subset1b["idx"])
diff = sorted(list(a.symmetric_difference(b)))
print("Symmetric diff: ", diff)
only_a = sorted(list(a.difference(b)))
print("Only in A: ", only_a)
only_b = sorted(list(b.difference(a)))
print("Only in B: ", only_b)
```
