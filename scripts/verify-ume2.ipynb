```python
import pandas as pd

dfs = []
names = ["ume-final", "ume2"]
for name in names:
    dfs.append(pd.read_json(f"../data/{name}.jsonl", lines=True))
df = pd.concat(dfs)
df
```

```python
df = df.replace("", pd.NA)
df["overall"] = pd.to_numeric(df["overall"], errors="coerce")
df["topic"] = pd.to_numeric(df["topic"], errors="coerce")
df["vocabulary"] = pd.to_numeric(df["vocabulary"], errors="coerce")
df["choices"] = pd.to_numeric(df["choices"], errors="coerce")


def filter_incomplete_responses(row):
    """Filter out rows with incomplete responses."""
    valid_values = [1, 2, 3]
    return (
        (row["overall"] in valid_values)
        & (row["topic"] in valid_values)
        & (row["vocabulary"] in valid_values)
        & (row["choices"] in valid_values)
    )

valid_df = df[df.apply(filter_incomplete_responses, axis=1)]
```

```python
counts = valid_df.groupby("idx").size()
items = counts[counts < 2].index.tolist()
valid_df[valid_df["idx"].isin(items)]
```
